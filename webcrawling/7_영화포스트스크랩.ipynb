```python
import requests
from bs4 import BeautifulSoup

url = "https://www.moviechart.co.kr/rank/realtime/index/image"
movie_ranking = requests.get(url)
soup = BeautifulSoup(movie_ranking.text, "html.parser")
soup
```

```python
##content > div.wArea.space > div.movieBox > ul > li:nth-child(1) > div > div.movie-title > h3 > a
#영화제목
#movie_list = soup.select("div.movie-title > h3 > a")
movie_list = soup.select("div.movieBox div.movie-title a")
for i in movie_list:
    print(i.text)
```
