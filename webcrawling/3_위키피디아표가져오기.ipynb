```python
##  requests 생성
import requests
from bs4 import BeautifulSoup

url = 'https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%95%84%EC%B9%B4%EB%8D%B0%EB%AF%B8_%EA%B5%AD%EC%A0%9C%EC%98%81%ED%99%94%EC%83%81_%EC%B6%9C%ED%92%88%EC%9E%91'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
table_element = soup.find('table')
tr_element = table_element.find_all('tr')
tr_element
```

```python
#요소 탐색
for row in tr_element:
    cells = row.find_all(['th', 'td'])
    #print(cells)
    for cell in cells:
        print(cell.text.strip(), end=' ')
```

```python
# csv 파일로 저장
import csv
file_name = 'wiki_table.csv'
with open(file_name, mode='w', newline='') as file:
    writer = csv.writer(file)
    for row in tr_element:
        cells = row.find_all(['th', 'td'])
        writer.writerow([cell.text.strip() for cell in cells]) #writer로 csv파일에 쓰기
```

```python
# csv 파일 읽기
with open(file_name, mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
