출력 예시
```
번호: 1, 이름: 김철수, 국어: 90, 영어: 80, 수학: 85, 총점: 255, 평균: 85.00
번호: 2, 이름: 박제동, 국어: 90, 영어: 85, 수학: 90, 총점: 265, 평균: 88.33
번호: 3, 이름: 홍길동, 국어: 80, 영어: 80, 수학: 80, 총점: 240, 평균: 80.00

반 평균  : 253.33
국어 평균: 86.67
영어 평균: 81.67
수학 평균: 85.00
```

````python

# 학생 데이터 초기화 : 최소 10명 이상으로 설정하세요.
#order는 석차

students = [
               {'num':'1','name':'김철수','kor':90,'eng':80,'math':85,'total':0,'avg':0.0,'order':0 },
               {'num':'2','name':'박제동','kor':90,'eng':85,'math':90,'total':0,'avg':0.0,'order':0 },
               {'num':'3','name':'홍길동','kor':80,'eng':80,'math':80,'total':0,'avg':0.0,'order':0 }
           ]

class_total = 0
class_avg = 0.0
kor_total = 0
eng_total = 0
math_total = 0
kor_avg = 0.0
eng_avg = 0.0
math_avg = 0.0

# 학생들의 성적 총점과 평균 및 반 총점과 과목별 총점을 구한다.
for s in students :
    #print(s['kor'], s['eng'] , s['math'])
    s['total'] = s['kor'] + s['eng'] + s['math']
    s['avg'] = s['total'] / 3
    
    #반 총점 계산
    class_total  = class_total + s['total']
    #과목별 총점
    kor_total  = kor_total + s['kor']
    eng_total  = eng_total + s['eng']
    math_total  = math_total + s['math']
    
    #print(s)

#순위 값 계속하기 위해서 새로운 리스트 작성
students_sorted = sorted(students, key= lambda s: s['total'], reverse=True )
#print(students_sorted)

#실제 데이터에 순위를 넣는다
for idx, student in enumerate(students_sorted, start=1) :
    student['order'] = idx
    
# 학생별 성적 처리 결과를 출력한다.
for s in students:
    print(f"번호: {s['num']}, 이름: {s['name']}, 국어: {s['kor']}, 영어: {s['eng']}, 수학: {s['math']}, 총점: {s['total']}, 평균: {s['avg']:.2f}")

# 반 평균 및 과목별 평균 을 계산한다.
class_avg = class_total/len(students)
kor_avg = kor_total/len(students)
eng_avg = eng_total/len(students)
math_avg = math_total/len(students)

#반 평균 및 각 과목별 평균을 출력한다.
print(f'반 평균  : {class_avg:.2f}')
print(f'국어 평균: {kor_avg:.2f}')
print(f'영어 평균: {eng_avg:.2f}')
print(f'수학 평균: {math_avg:.2f}')

# 팀당 요일구하기 프로그램의 순서도, 성적처리시스템

````

## 함수화

````python
import random

# 학생 데이터(점수)를 생성하는 함수
def generate_student_data(names):
    students = []
    for i, name in enumerate(names, start=1):
        student = {
            'num': str(i),
            'name': name,
            'kor': random.randint(50, 100),  # 국어 점수 50~100점 사이
            'eng': random.randint(50, 100),  # 영어 점수 50~100점 사이
            'math': random.randint(50, 100)  # 수학 점수 50~100점 사이
        }
        students.append(student)
    return students
````

````python
def calculate_student_scores(students):
    """학생들의 총점과 평균을 계산"""
    for s in students:
        s['total'] = s['kor'] + s['eng'] + s['math']
        s['avg'] = s['total'] / 3

def calculate_class_totals(students):
    """반의 총점 및 과목별 총점을 계산"""
    kor_total = sum(s['kor'] for s in students)
    eng_total = sum(s['eng'] for s in students)
    math_total = sum(s['math'] for s in students)
    class_total = sum(s['total'] for s in students)
    return kor_total, eng_total, math_total, class_total

def calculate_class_averages(kor_total, eng_total, math_total, class_total, num_students):
    """반 평균 및 과목별 평균을 계산"""
    kor_avg = kor_total / num_students
    eng_avg = eng_total / num_students
    math_avg = math_total / num_students
    class_avg = class_total / num_students
    return kor_avg, eng_avg, math_avg, class_avg

def assign_student_ranks(students):
    """총점을 기준으로 학생들에게 석차를 부여"""
    students_sorted = sorted(students, key=lambda s: s['total'], reverse=True)
    for idx, student in enumerate(students_sorted, start=1):
        student['order'] = idx

def print_student_scores(students):
    """학생별 성적을 출력"""
    for s in students:
        print(f"번호: {s['num']}, 이름: {s['name']}, 국어: {s['kor']}, 영어: {s['eng']}, 수학: {s['math']}, "
              f"총점: {s['total']}, 평균: {s['avg']:.2f}, 석차: {s['order']}")

def print_class_averages(class_avg, kor_avg, eng_avg, math_avg):
    """반 평균 및 과목별 평균을 출력"""
    print(f'\n반 평균  : {class_avg:.2f}')
    print(f'국어 평균: {kor_avg:.2f}')
    print(f'영어 평균: {eng_avg:.2f}')
    print(f'수학 평균: {math_avg:.2f}')

# 메인 함수
def main():
    """성적 처리 시스템 메인 함수"""
    
    # 학생 데이터 초기화 : 최소 10명 이상으로 설정하세요.
    student_names = ['김철수', '박제동', '홍길동', '이영희', '최지우', '강호동', '유재석', '신동엽', '이승기', '정형돈']
    students = generate_student_data(student_names)

    # 학생별 성적 계산
    calculate_student_scores(students)

    # 학생 석차 계산
    assign_student_ranks(students)

    # 학생별 성적 출력
    print_student_scores(students)

    # 반의 총점 및 과목별 총점 계산
    kor_total, eng_total, math_total, class_total = calculate_class_totals(students)

    # 반 평균 및 과목별 평균 계산
    kor_avg, eng_avg, math_avg, class_avg = calculate_class_averages(kor_total, eng_total, math_total, class_total, len(students))

    # 반 평균 및 과목별 평균 출력
    print_class_averages(class_avg, kor_avg, eng_avg, math_avg)

# 프로그램 실행


main()
````

## Student 및 Classroom 클래스에서 캡슐화 적용

````python
import random

class Student:
    """학생 클래스: 각 학생의 정보를 관리"""
    
    def __init__(self, num, name, kor, eng, math):
        self.__num = num  # 학생 번호 (비공개)
        self.__name = name  # 학생 이름 (비공개)
        self.__kor = kor  # 국어 점수 (비공개)
        self.__eng = eng  # 영어 점수 (비공개)
        self.__math = math  # 수학 점수 (비공개)
        self.__total = 0  # 총점 (비공개)
        self.__avg = 0.0  # 평균 (비공개)
        self.__order = 0  # 석차 (비공개)

    def calculate_scores(self):
        """총점과 평균을 계산"""
        self.__total = self.__kor + self.__eng + self.__math
        self.__avg = self.__total / 3

    # Getter methods (캡슐화된 비공개 속성에 접근하기 위해 제공)
    def get_num(self):
        return self.__num

    def get_name(self):
        return self.__name

    def get_kor(self):
        return self.__kor

    def get_eng(self):
        return self.__eng

    def get_math(self):
        return self.__math

    def get_total(self):
        return self.__total

    def get_avg(self):
        return self.__avg

    def get_order(self):
        return self.__order

    # Setter methods (비공개 속성에 값을 설정할 때 사용)
    def set_order(self, order):
        self.__order = order


class Classroom:
    """반 전체 학생 데이터를 관리하고 처리하는 클래스"""

    def __init__(self, student_names):
        self.__students = self.__generate_student_data(student_names)

    def __generate_student_data(self, names):
        """비공개 메서드: 학생 데이터를 생성"""
        students = []
        for i, name in enumerate(names, start=1):
            student = Student(
                num=str(i),
                name=name,
                kor=random.randint(50, 100),
                eng=random.randint(50, 100),
                math=random.randint(50, 100)
            )
            students.append(student)
        return students

    def calculate_student_scores(self):
        """모든 학생의 총점과 평균을 계산"""
        for student in self.__students:
            student.calculate_scores()

    def assign_student_ranks(self):
        """총점을 기준으로 학생들의 석차를 부여"""
        students_sorted = sorted(self.__students, key=lambda s: s.get_total(), reverse=True)
        for idx, student in enumerate(students_sorted, start=1):
            student.set_order(idx)

    def calculate_class_totals(self):
        """반의 총점 및 과목별 총점을 계산"""
        kor_total = sum(s.get_kor() for s in self.__students)
        eng_total = sum(s.get_eng() for s in self.__students)
        math_total = sum(s.get_math() for s in self.__students)
        class_total = sum(s.get_total() for s in self.__students)
        return kor_total, eng_total, math_total, class_total

    def calculate_class_averages(self, kor_total, eng_total, math_total, class_total):
        """반 평균 및 과목별 평균을 계산"""
        num_students = len(self.__students)
        kor_avg = kor_total / num_students
        eng_avg = eng_total / num_students
        math_avg = math_total / num_students
        class_avg = class_total / num_students
        return kor_avg, eng_avg, math_avg, class_avg

    def print_student_scores(self):
        """학생별 성적을 출력"""
        for student in self.__students:
            print(f"번호: {student.get_num()}, 이름: {student.get_name()}, 국어: {student.get_kor()}, 영어: {student.get_eng()}, "
                  f"수학: {student.get_math()}, 총점: {student.get_total()}, 평균: {student.get_avg():.2f}, 석차: {student.get_order()}")

    def print_class_averages(self, class_avg, kor_avg, eng_avg, math_avg):
        """반 평균 및 과목별 평균을 출력"""
        print(f'\n반 평균  : {class_avg:.2f}')
        print(f'국어 평균: {kor_avg:.2f}')
        print(f'영어 평균: {eng_avg:.2f}')
        print(f'수학 평균: {math_avg:.2f}')


# 메인 함수
def main():
    """성적 처리 시스템 메인 함수"""
    
    # 학생 이름 리스트
    student_names = ['김철수', '박제동', '홍길동', '이영희', '최지우', '강호동', '유재석', '신동엽', '이승기', '정형돈']
    
    # Classroom 객체 생성
    classroom = Classroom(student_names)

    # 학생별 성적 계산
    classroom.calculate_student_scores()

    # 학생 석차 계산
    classroom.assign_student_ranks()

    # 학생별 성적 출력
    classroom.print_student_scores()

    # 반의 총점 및 과목별 총점 계산
    kor_total, eng_total, math_total, class_total = classroom.calculate_class_totals()

    # 반 평균 및 과목별 평균 계산
    kor_avg, eng_avg, math_avg, class_avg = classroom.calculate_class_averages(kor_total, eng_total, math_total, class_total)

    # 반 평균 및 과목별 평균 출력
    classroom.print_class_averages(class_avg, kor_avg, eng_avg, math_avg)

# 프로그램 실행
if __name__ == '__main__':
    main()
````
