```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10, 100)
y = np.sin(x)

plt.plot(x,y)
plt.title('사인 곡선')
plt.xlabel('시간')
plt.ylabel('진폭')
plt.show()
```
